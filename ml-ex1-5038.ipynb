Implementation of Python Basic Libraries such as Statistics, Math, Numpy and Scipy

Usage of methods such as floor(), ceil(), sqrt(), isqrt(), gcd() etc

```python
import math
```

```python
print(math.ceil(5.3))
print(math.floor(5.4))
print(math.sqrt(16))
print(math.isqrt(167))
print(math.gcd(12,8))
```

Usage of attributes of array such as ndim, shape, size, methods such as sum(), mean(), sort(), sin() etc.

```python
import numpy as np
arr = np.array([[1,2,3],[6,5,4]])
print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.sum())
print(arr.mean())
print(np.sort(arr))
print("Sine values of each number of the array are \n",np.sin(arr))
```

Usage of methods such as det(), eig() etc

```python
matrix = np.array([[1,2,4],[9,7,5],[3,4,5]])
print("Determinant of the matrix is ",np.linalg.det(matrix))
eigenvalues,eigenvectors = np.linalg.eig(matrix)
print("Eigen values are ",eigenvalues)
print("Eigen vectors are \n",eigenvectors)
```

Consider a list datatype(1D) then reshape it into2D, 3D matrix using numpy

```python
lst = list( range(12))
array = np.array(lst)
arr_2d = array.reshape((3,4))
print("The 1D array into 2D array as \n",arr_2d)
arr_3d = array.reshape((2,2,3))
print("The 1D array into 2D array as \n",arr_3d)
```

Generater and matrices using numpy

```python
zero_matrix = np.zeros((3,4))
ones_matrix = np.ones((2,2))
identity_matrix = np.eye(3)
random_matrix = np.random.rand(3,3)

print("Zero matrix \n",zero_matrix)
print("One matrix \n",ones_matrix)
print("Identity matrix \n",identity_matrix)
print("Random matrix \n",random_matrix)
```

 Find the determinant of a matrix using scipy

```python
from scipy.linalg import det,eig
mat = np.array([[1, 2],[3, 4]])
determinant = det(mat)
print(determinant)
```

```python
mat = np.array([[5, 4],
                [1, 2]])
eigenvalues, eigenvectors = eig(mat)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

Define Series objects with individual indices

```python
data = [100,200,300]
indices = ['a','b','c']
customIndexArray = pd.Series(data,index=indices)
print("The array with custom indices \n",customIndexArray)
```

 Access single value of a series

```python
print(customIndexArray['a'])
print(customIndexArray['b'])
```
